```python
import pandas as pd

train = pd.read_csv('data/train.csv', encoding='utf-8-sig')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
from scipy import stats
```

```python
# 변환전 
plt.figure(figsize=(16, 6))
plt.subplot(1, 3, 1)
plt.hist(train['전력소비량(kWh)'], bins=50, alpha=0.7, edgecolor='black')
plt.title('전력소비량 분포')
plt.xlabel('전력소비량')
plt.ylabel('빈도')

# -------------------------------------------------------------

# 변환후
if train['전력소비량(kWh)'].min() >= 0:
    train['전력소비량_log'] = np.log1p(train['전력소비량(kWh)'])
    transformation_method = "log1p 변환"

plt.figure(figsize=(8, 6))
plt.hist(train['전력소비량_log'], bins=50, alpha=0.7, color='orange', edgecolor='black')
plt.title('로그변환 후 전력소비량 분포')
plt.xlabel('log(전력소비량)')
plt.ylabel('빈도')

# QQ-plot으로 정규성 확인
plt.figure(figsize=(8, 6))
stats.probplot(train['전력소비량_log'], dist="norm", plot=plt)
plt.title('QQ-plot (정규성 검정)')
plt.tight_layout()
plt.show()
```
